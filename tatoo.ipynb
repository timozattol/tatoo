### Try to show how difficult it is to find the right mapping with a given cut

Randomly assign one letter to each letter of the passord, as if we discovered how to cut the huffmann code in segments and we then want to find the right letter mapping, via brute force, using english words

```python
words = ["ABCDED", "GA", "HID", "JDAH"]
```

```python
words_corpus = set([
    "SOLENE",
    "IS",
    "THE",
    "BEST",
    "COOL",
    "WEST",
    "BLEU",
    "WRONG",
    "YOP",
    "SWAG",
    "NICE",
    "RIGHTS",
    "WRONGS",
    "EASY",
    "TO",
    "NO",
    "YES",
    "WILLYS",
])
```

```python
def conflicting(dict1, dict2):
    """Return True if one key from dict1 is present in dict2 but the values differ"""
    for key, value in dict1.items():
        if key in dict2 and dict2[key] != value:
            return True
        
    return False
    

def print_possibilities(constructed_sentence, sequences_left, current_mapping, words_corpus, max_level, current_level=0):
    if len(sequences_left) == 0 or current_level == max_level:
        print(constructed_sentence)
        return
    
    next_sequence = sequences_left[0]
    
    # Try all possible words for next word
    for possible_word in words_corpus:
        
        if len(possible_word) != len(next_sequence):
            continue
        # Mapping from sequence letter to possible words
        new_mapping = dict(zip(next_sequence, possible_word))
        if conflicting(new_mapping, current_mapping):
            continue
        
        merged_mapping = current_mapping.copy()
        merged_mapping.update(new_mapping)

        new_sentence = "{} {}".format(constructed_sentence, possible_word)
        new_sequences_left = sequences_left[1:]

        print_possibilities(new_sentence, new_sequences_left, merged_mapping, words_corpus, max_level, current_level + 1)

print_possibilities("", list(words), {}, words_corpus, max_level=4)
```
